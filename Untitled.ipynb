```python
from pandas import read_csv, merge, concat
%matplotlib inline
```

```python
fn0 = 'SIXcsv/SXitd_SPICHA_2016-11-09_163212.csv'
fn1 = 'SIXcsv/SXitd_SPI_2016-11-09_163212.csv'
fn2 = 'SIXcsv/SXitd_CHSPI_2016-11-09_163212.csv'
```

```python
etf2.head(1)
```

```python
etf1 = read_csv(fn0, encoding='latin1', delimiter=';', skiprows=2, usecols=[0,1,2], index_col=0)
etf1.columns = ['etf1','etf1Vol']
spi = read_csv(fn1, encoding='latin1', delimiter=';', skiprows=2, usecols=[0,1,2], index_col=0)
del spi['Volume']
spi.columns = ['spi']
etf2 = read_csv(fn2, encoding='latin1', delimiter=';', skiprows=2, usecols=[0,1,2], index_col=0)
etf2.columns = ['etf2','etf2Vol']
```

```python
etf1.etf1.plot(figsize=(10,2))
```

```python
etf2.etf2.plot(figsize=(10,2))
```

```python
spi.spi.plot(figsize=(10,2))
```

```python
spid = concat([spi,etf1,etf2], join='outer')
```

```python
spid.spi.dropna().plot()
```

```python
fondsdiff = lambda fonds, fak : (spid.spi.interpolate()-fak * spid[fonds].dropna()).pow(2).sum()
```

```python
fondsdiff('etf1', 1)
```

```python
for etf in ['etf1', 'etf2']:
    fak = 5
    inc = 50
    diffold = fondsdiff(etf, fak)
    fak = fak + inc
    for i in range(100):
        diff = fondsdiff(etf, fak)
        if diff > diffold:
            inc = -inc/2
            fak = fak + inc
        else:
            fak = fak + inc
        diffold = diff
    print(fak,inc)
    spid[etf] *= fak
        
    
```

```python
spid.etf1.plot()
spid.etf2.plot()
spid.spi.plot()
```
