## read all SIX csv files and create pandas datafram

```python
from pandas import read_csv, MultiIndex, concat, read_excel, NaT
from collections import namedtuple
from datetime import time, datetime
import matplotlib.pyplot as plt
import codecs
import os

%matplotlib inline
```

```python
def get_pfad():
    '''
    returns path for different files as needed per PC
    '''

    Context = namedtuple('contexts', 'pfadKGAST, pfadSIX')
    pfade = ['c:/users/gfi 64bit/',
             'c:/users/gerar/',
             'c:/users/gérard/']
    try:
        pfad = [pfad for pfad in pfade if os.path.exists(pfad)][0]
    except:
        pfad = ''
        raise FileNotFoundError('Zuerst gültigen Pfad definieren!')
    pfad += 'dropbox/KGAST Daten/'
    context = Context(pfad + 'KGASTxls/', pfad + 'SIXcsv/')
    return context

def to_float(x):
    ''' convert x to float or NaN '''
    try:
        return float(str(x).replace("'",""))
    except:
        return float('NaN')
    
def to_datetime(datum):
    ''' convert x to datetime or NaT '''
    try:
        return datetime(*[int(i) for i in datum.split('-')]).date()
    except:
        return NaT
```

```python
p = get_pfad()
print(p.pfadKGAST, ' -- ',p.pfadKGAST)
fnames = os.listdir(p.pfadKGAST)
fnames =[fname for fname in fnames if 'KGAST' in fname] # keep intraday
print(fnames[1])

flist = []
#fnames = fnames[:100] # short read for testing
for fn in fnames:
    flist.append(read_excel(p.pfadKGAST + fn))

df = concat(flist).drop_duplicates()
df.info(verbose=True)
dff = df.copy()
```

```python
df = dff.copy()
df.head(2)
```

```python
df.AUSGABEPREIS = df.AUSGABEPREIS.map(to_float)
df.RUECKNAHMEPREIS = df.RUECKNAHMEPREIS.map(to_float)
df.AUFGEL_ERTRAG = df.AUFGEL_ERTRAG.map(to_float)

df.ABSCHLUSSTAG = df.ABSCHLUSSTAG.map(to_datetime)
df.VALUTA = df.VALUTA.map(to_datetime)
df.AENDERUNGSDAT = df.AENDERUNGSDAT.map(to_datetime)
```
